### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1=tienda['Precio'].sum()
ingreso_tienda1
```

```python
ingreso_tienda2=tienda2['Precio'].sum()
ingreso_tienda2
```

```python
ingreso_tienda3=tienda3['Precio'].sum()
ingreso_tienda3
```

```python
ingreso_tienda4=tienda4['Precio'].sum()
ingreso_tienda4
```

**Grafico de facturación por tienda**

En el siguiente grafico se muestra la comparación de los valores de facturación de las 4 tiendas

```python
data = {'Tienda': ['Tienda1', 'Tienda2', 'Tienda3', 'Tienda4'],
        'Ventas': [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]}
df = pd.DataFrame(data)

# 2. Crear el gráfico de barras
plt.bar(df['Tienda'], df['Ventas'], color='skyblue')

# 3. Añadir títulos y etiquetas
plt.title('Ventas por tienda')
plt.xlabel('Tienda')
plt.ylabel('Ventas ($)')

# 4. Mostrar el gráfico
plt.show()
```

# 3. Calificación promedio de la tienda


```python
promedio1 = tienda['Calificación'].mean()
promedio1
```

```python
promedio2 = tienda2['Calificación'].mean()
promedio2
```

```python
promedio3 = tienda3['Calificación'].mean()
promedio3
```

```python
promedio4 = tienda4['Calificación'].mean()
promedio4
```

```python
#Grafico de evaluación por tienda
# Datos para el gráfico
etiquetas = ['Tienda1', 'Tienda2', 'Tienda3', 'Tienda4']
cantidades = [promedio1, promedio2, promedio3, promedio4]
colores = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']


# Crear el gráfico
plt.pie(cantidades,
        labels=etiquetas,
        autopct='%1.1f%%',   # Muestra el porcentaje con un decimal
        colors=colores,
        startangle=140)      # Ángulo de inicio

plt.title("Grafico de evaluacion por tienda")
plt.axis('equal')  # Asegura que el gráfico sea un círculo perfecto
plt.show()
```
